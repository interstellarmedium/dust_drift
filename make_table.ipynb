### merge tables of data and results and write out in latex format for the paper

```python
import numpy as np
import pandas as pd

%matplotlib inline
```

```python
colnames = ['Source', 'dist', 'lambda', 'Rmm90', 'RCO90', 'Mdust', 'Lstar', 'Teff', 'Mstar', 'age', 'Ref']
df = pd.read_csv('ALL_combined_table.txt', skiprows=2, header=0, sep='\s+', names=colnames)
```

```python
colnames = ['Source', 't_star', 't_drift_min', 't_drift_max', 'M_min', 'M_max']
df2 = pd.read_csv('drift_fit_results.txt', skiprows=1, header=0, sep='\s+', names=colnames)
```

```python
df_merge = df[['Source', 'age', 'Lstar', 'Teff', 'Mstar', 'Mdust', 'Rmm90', 'lambda', 'RCO90']]
```

```python
df_merge = df_merge.merge(df2[['Source', 't_drift_min', 't_drift_max', 'M_max']], on='Source')
```

```python
df_merge
```

```python
f = open("table1.tex", 'w')
f.write(r"Source & $t_\ast$ & $L_\ast$ & $T_{\rm eff}$ & $M_\ast$ & $M_{\rm dust}$ & $R_{\rm mm, 90}$ & $\lambda$ & $R_{\rm CO,90}$ & $t_{\rm drift,1}$ & $t_{\rm drift,2}$ & $M_{\rm disk,2}$ \\")
f.write("\n")
f.write(r" & (Myr) & ($L_\odot$) & (K) & ($M_\odot$) & ($M_{\rm oplus}$) & (au) & (mm) & (au) & (Myr) & (Myr) & ($M_\odot$) \\")
f.write("\n")
for i in range(len(df_merge)):
    f.write(f"{df_merge['Source'][i]:17s} & {df_merge['age'][i]:4.1f} & {df_merge['Lstar'][i]:5.2f} \
& {df_merge['Teff'][i]:4g} & {df_merge['Mstar'][i]:5.2f} & {df_merge['Mdust'][i]:3g} \
& {df_merge['Rmm90'][i]:3d} & {df_merge['lambda'][i]:3.1f} & {df_merge['RCO90'][i]:4d} \
& {df_merge['t_drift_min'][i]:4.2f} & {df_merge['t_drift_max'][i]:5.2f} & {df_merge['M_max'][i]:5.3f} \\\\")
    f.write("\n")
f.close()
```
